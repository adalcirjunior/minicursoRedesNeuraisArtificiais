```python
#Importandos os dados de um arquivo *.xlsx
dados = pd.read_excel('https://github.com/adalcirjunior/minicursoRedesNeuraisArtificiais/raw/main/experimentoMola.xlsx')
```

```python
#Eliminando a coluna cargas
dados2 = dados.drop('cargas', axis=1)
```

```python
#Eliminando a coluna Lo
dados3 = dados2.drop('Lo', axis=1)
```

```python
#Elaborando um gráfico de dispersão entre a deformação e a força
plt.scatter(dados3.x,dados3.forca)
plt.xlabel('Deformação (mm)')
plt.ylabel('Força (N)')
plt.title('Rigidez da mola')
plt.show()
```

```python
#Determinando o coeficiente de correlação
#Métodos de correlação: pearson (padrão), spearman e kendall
coeficienteDeCorrelacao = dados3['x'].corr(dados3['forca'], method = 'pearson')
print(coeficienteDeCorrelacao)
```

```python
#Determinando o coeficiente de determinação
coeficienteDeDeterminacao = coeficienteDeCorrelacao**2
print(coeficienteDeDeterminacao)
```

```python
#Selecionando os dados de x (coluna x)
x = dados3.iloc[:, 2:3].values

#Selecionando os dados de y (coluna forca)
y = dados3.iloc[:, 0:1].values
```

```python
#Importando a biblioteca MLPRegressor
from sklearn.neural_network import MLPRegressor
```

```python
#Criando a rede reural artificial
regressor = MLPRegressor(hidden_layer_sizes=(26,26), activation='relu', verbose=True, batch_size=1, solver='lbfgs', random_state = 12)
```

```python
#Treinando a rede neural artificial
regressor.fit(x, y)
```

```python
#Verificando o número de camadas da rede neural artificial
regressor.n_layers_
```

```python
#verificando o coeficiente de determinação
regressor.score(x, y)
```

```python
deformacao = 0.06
```

```python
previsao_forca = regressor.predict(np.array(deformacao).reshape(1,1))
```
